### Group the marketing qualified leads by Month, Year and Origin (marketing channel)

```python
path1=r'C:\Users\user\Desktop\Anastasios\Personal\Data_Science\RBS_MSc_Data_Science\Module 5 _Presentation\Olist_HW\olist_marketing_qualified_leads_dataset.csv'
```

```python
#insert the csv file with the marketing data
df1=pd.read_csv(path1)
df1.head()
```

```python
#convert the "first_contact_date" to datetime data type
df1.first_contact_date=pd.to_datetime(df1.first_contact_date)
```

```python
#create 2 new columns with the month and the year
df1['month']=df1['first_contact_date'].dt.month_name()
df1['year']=df1['first_contact_date'].dt.year
df1.head()
```

```python
#group by month, year, origin with the total number of qualified leads
df2=df1.groupby(['month', 'year', 'origin'])[['origin']].count()
df2.head()
```

```python
#rename the column name
df2.columns =['Num_leads']
df2.head()
```

```python
df2.reset_index(inplace=True)
df2.head()
```
